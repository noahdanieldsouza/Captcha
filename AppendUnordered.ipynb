```python
import csv
from google.colab import files

uploaded = files.upload()

with open(list(uploaded.keys())[0], newline='', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    read_data_columns = reader.fieldnames


    order = {}
    for row in reader:
      key = (row['state'], row['county_name'], row['county_fips'])
      if key not in order:
        order[key] = {
         "state": row['state'],
         "county": row['county_name'], "county_fips": row['county_fips']}


    print (order.values())
```

gets the order the counties should be added on based on their current position
